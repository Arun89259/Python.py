```python
# Q1 Print numbers from 1 to 10

i = 1
while i <= 10:
    print(i)
    i += 1
```

```python
# Q2 Print numbers from 10 to 1

i = 10
while i >= 1:
    print(i)
    i -= 1
```

```python
# Q3 Even numbers between 1 and 20

i = 1
while i <= 20:
    if i % 2 == 0:
        print(i)
    i += 1
```

```python
# Q4 Odd numbers between 1 and 20

i = 1
while i <= 20:
    if i % 2 != 0:
        print(i)
    i += 1
```

```python
# Q5 Numbers divisible by 5 (1–50)

i = 1
while i <= 50:
    if i % 5 == 0:
        print(i)
    i += 1
```

```python
# Q6 Numbers divisible by both 3 and 5

i = 1
while i <= 50:
    if i % 3 == 0 and i % 5 == 0:
        print(i)
    i += 1
```

```python
# Q7 Numbers NOT divisible by 3 (1–100)

i = 1
while i <= 100:
    if i % 3 != 0:
        print(i)
    i += 1
```

```python
# Q8 Numbers divisible by 7 (1–100)

i = 1
while i <= 100:
    if i % 7 == 0:
        print(i)
    i += 1
```

```python
# Q9 Skip multiples of 3

i = 1
while i <= 30:
    if i % 3 == 0:
        i += 1
        continue
    print(i)
    i += 1
```

```python
# Q10 Stop when number becomes 15

i = 1
while i <= 20:
    if i == 15:
        break
    print(i)
    i += 1
```

```python
# Q11 Even or Odd label

i = 1
while i <= 50:
    if i % 2 == 0:
        print(i, "Even")
    else:
        print(i, "Odd")
    i += 1
```

```python
# Q12 Divisible by 3 but not by 5

i = 1
while i <= 100:
    if i % 3 == 0 and i % 5 != 0:
        print(i)
    i += 1
```

```python
# Q13 Divisible by 5 but not by 3

i = 1
while i <= 100:
    if i % 5 == 0 and i % 3 != 0:
        print(i)
    i += 1
```

```python
# Q14 Divisible by 3 OR 5

i = 1
while i <= 100:
    if i % 3 == 0 or i % 5 == 0:
        print(i)
    i += 1
```

```python
# Q15 Divisible by both 2 and 3

i = 1
while i <= 100:
    if i % 2 == 0 and i % 3 == 0:
        print(i)
    i += 1
```
