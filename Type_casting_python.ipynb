# Type Casting 


| Function    | Use                              |
| ----------- | -------------------------------- |
| `int()`     | kisi value ko integer banata hai |
| `float()`   | decimal number banata hai        |
| `str()`     | string (text) banata hai         |
| `bool()`    | True / False banata hai          |
| `complex()` | complex number banata hai        |



## 1. String (str) Casting

```python
# Int → string
print(str(10))          # ✅ "10"

# Float → string
print(str(10.5))        # ✅ "10.5"

# Bool → string
print(str(True))        # ✅ "True"

# Complex → string
print(str(2+3j))        # ✅ "(2+3j)"
```

## 2. Integer (int) Casting

```python
# String → int
print(int("10"))        # ✅ 10
# print(int("10.5"))    # ❌ Error
# print(int("hello"))   # ❌ Error

# Float → int
print(int(10.9))        # ✅ 10

# Bool → int
print(int(True))        # ✅ 1
print(int(False))       # ✅ 0

# Complex → int
# print(int(2+3j))      # ❌ Error
```

## 3. Float (float) Casting

```python
# String → float
print(float("10"))      # ✅ 10.0
print(float("10.5"))    # ✅ 10.5
# print(float("hello")) # ❌ Error

# Int → float
print(float(10))        # ✅ 10.0

# Bool → float
print(float(True))      # ✅ 1.0
print(float(False))     # ✅ 0.0

# Complex → float
# print(float(2+3j))    # ❌ Error
```

## 4. Boolean (bool) Casting

```python
# Int → bool
print(bool(10))         # ✅ True
print(bool(0))          # ❌ False

# Float → bool
print(bool(10.5))       # ✅ True
print(bool(0.0))        # ❌ False

# String → bool
print(bool("Hello"))    # ✅ True
print(bool(""))         # ❌ False

# Complex → bool
print(bool(2+3j))       # ✅ True
print(bool(0+0j))       # ❌ False
```

## 5. Complex (complex) Casting

```python
# Int → complex
print(complex(10))        # ✅ (10+0j)

# Float → complex
print(complex(10.5))      # ✅ (10.5+0j)

# Two numbers → complex
print(complex(2, 3))      # ✅ (2+3j)

# String → complex
print(complex("3+5j"))    # ✅ (3+5j)
print(complex("10"))      # ✅ (10+0j)
# print(complex("hello")) # ❌ Error
```
