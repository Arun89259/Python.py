# MIXED QUESTIONS

```python
# Mixed 1. Take a number and print its type, binary, octal, hex.
n = 10
print(type(n))
print(bin(n))
print(oct(n))
print(hex(n))
```

```python
# Mixed 2. Print Decimal, Binary, Octal, Hex for 10.
n = 10
print("Decimal:", n)
print("Binary:", bin(n))
print("Octal:", oct(n))
print("Hex:", hex(n))
```

```python
# Mixed 3. Convert int → binary → back to int and check type.
n = 10
b = bin(n)
n2 = int(b, 2)
print(n2, type(n2))
```

```python
# Mixed 4. Print a number in all formats using one program.
n = 25
print("Decimal:", n)
print("Binary:", bin(n))
print("Octal:", oct(n))
print("Hex:", hex(n))
```
