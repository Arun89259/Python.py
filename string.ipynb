```python
# Print each character of a string
# Input: "hello"
# Output:
# h
# e
# l
# l
# o

text = "hello"

for ch in text:
    print(ch)
```

```python
# Count total characters (without using len)
# Input: "python"
# Output: 6

text = "python"

count = 0
for ch in text:
    count += 1

print(count)
```

```python
# Count occurrence of a character
# Input: String = "banana", Character = 'a'
# Output: 3

text = "banana"
char = "a"

count = 0
for ch in text:
    if ch == char:
        count += 1

print(count)
```

```python
# Print characters at even index positions
# Input: "abcdef"
# Output: a c e

text = "abcdef"

for i in range(len(text)):
    if i % 2 == 0:
        print(text[i], end=" ")
```

```python
# Print characters at odd index positions
# Input: "abcdef"
# Output: b d f

text = "abcdef"

for i in range(len(text)):
    if i % 2 != 0:
        print(text[i], end=" ")
```

```python
# Reverse a string (without slicing)
# Input: "hello"
# Output: olleh

text = "hello"
rev = ""

for ch in text:
    rev = ch + rev

print(rev)
```

```python
# Count vowels
# Input: "education"
# Output: 5

text = "education"

count = 0
for ch in text:
    if ch in "aeiou":
        count += 1

print(count)
```

```python
# Count consonants
# Input: "hello"
# Output: 3

text = "hello"

count = 0
for ch in text:
    if ch.isalpha() and ch not in "aeiou":
        count += 1

print(count)
```

```python
# Count digits in a string
# Input: "a1b2c3"
# Output: 3

text = "a1b2c3"

count = 0
for ch in text:
    if ch.isdigit():
        count += 1

print(count)
```

```python
# Count spaces in a string
# Input: "hi hello how"
# Output: 2

text = "hi hello how"

count = 0
for ch in text:
    if ch == " ":
        count += 1

print(count)
```

```python
# Replace spaces with '-'
# Input: "hi hello"
# Output: hi-hello

text = "hi hello"
result = ""

for ch in text:
    if ch == " ":
        result += "-"
    else:
        result += ch

print(result)
```

```python
# Find first occurrence index of a character
# Input: "banana", 'a'
# Output: 1

text = "banana"
char = "a"

for i in range(len(text)):
    if text[i] == char:
        print(i)
        break
```

```python
# Find all indexes of a character
# Input: "banana", 'a'
# Output: 1 3 5

text = "banana"
char = "a"

for i in range(len(text)):
    if text[i] == char:
        print(i, end=" ")
```

```python
# Check if string is palindrome
# Input: "madam"
# Output: Palindrome

text = "madam"
rev = ""

for ch in text:
    rev = ch + rev

if text == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
# Find length of string manually
# Input: "data"
# Output: 4

text = "data"

count = 0
for ch in text:
    count += 1

print(count)
```

```python
# Count uppercase letters
# Input: "HeLLo"
# Output: 3

text = "HeLLo"

count = 0
for ch in text:
    if ch.isupper():
        count += 1

print(count)
```

```python
# Count lowercase letters
# Input: "HeLLo"
# Output: 2

text = "HeLLo"

count = 0
for ch in text:
    if ch.islower():
        count += 1

print(count)
```

```python
# Remove all vowels
# Input: "education"
# Output: dctn

text = "education"
result = ""

for ch in text:
    if ch not in "aeiou":
        result += ch

print(result)
```

```python
# Remove all digits
# Input: "a1b2c3"
# Output: abc

text = "a1b2c3"
result = ""

for ch in text:
    if not ch.isdigit():
        result += ch

print(result)
```

```python
# Print only unique characters
# Input: "programming"
# Output: progamin

text = "programming"
result = ""

for ch in text:
    if ch not in result:
        result += ch

print(result)
```

```python
# Convert lowercase to uppercase manually
# Input: "abc"
# Output: ABC

text = "abc"
result = ""

for ch in text:
    result += chr(ord(ch) - 32)

print(result)
```

```python
# Shift each character by 1
# Input: "abc"
# Output: bcd

text = "abc"
result = ""

for ch in text:
    result += chr(ord(ch) + 1)

print(result)
```

```python
# Find longest repeating character sequence
# Input: "aaabbccccd"
# Output: cccc

text = "aaabbccccd"

max_seq = ""
current = text[0]

for i in range(1, len(text)):
    if text[i] == text[i-1]:
        current += text[i]
    else:
        if len(current) > len(max_seq):
            max_seq = current
        current = text[i]

if len(current) > len(max_seq):
    max_seq = current

print(max_seq)
```

```python
# Check if string contains only alphabets
# Input: "hello"
# Output: Valid

text = "hello"

if text.isalpha():
    print("Valid")
else:
    print("Invalid")
```

```python
# Check if string contains only digits
# Input: "12345"
# Output: Valid

text = "12345"

if text.isdigit():
    print("Valid")
else:
    print("Invalid")
```

```python
# Merge two strings alternately
# Input: "abc", "123"
# Output: a1b2c3

s1 = "abc"
s2 = "123"

result = ""

for i in range(len(s1)):
    result += s1[i] + s2[i]

print(result)
```

```python
# Check if two strings are equal (without using ==)
# Input: "hello", "hello"
# Output: Equal

s1 = "hello"
s2 = "hello"

equal = True

if len(s1) != len(s2):
    equal = False
else:
    for i in range(len(s1)):
        if s1[i] != s2[i]:
            equal = False
            break

if equal:
    print("Equal")
else:
    print("Not Equal")
```
