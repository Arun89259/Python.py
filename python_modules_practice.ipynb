## Random Module

### random

```python
import random
random.random()  # Example output: 0.63
```

### randint

```python
import random
random.randint(1,10)  # Example output: 7
```

### randrange

```python
import random
random.randrange(1,10,2)  # Example output: 5
```

### choice

```python
import random
random.choice(['A','B','C'])  # Example output: 'B'
```

### choices

```python
import random
random.choices([1,2,3],k=3)  # Example output: [2,1,3]
```

### shuffle

```python
import random
nums=[1,2,3,4]
random.shuffle(nums)
nums  # Example output: [3,1,4,2]
```

### sample

```python
import random
random.sample([1,2,3,4,5],2)  # Example output: [4,2]
```

### uniform

```python
import random
random.uniform(1,5)  # Example output: 3.42
```

## Datetime Module

### datetime

```python
import datetime
datetime.datetime.now()
```

### date

```python
import datetime
datetime.date(2026,3,26)
```

### time

```python
import datetime
datetime.time(10,30,20)
```

### timedelta

```python
from datetime import datetime, timedelta
datetime.now()+timedelta(days=5)
```

### now

```python
from datetime import datetime
datetime.now()
```

### today

```python
from datetime import date
date.today()
```

### strftime

```python
from datetime import datetime
datetime.now().strftime('%d-%m-%Y')
```

### strptime

```python
from datetime import datetime
datetime.strptime('26-03-2026','%d-%m-%Y')
```

### timestamp

```python
from datetime import datetime
datetime.now().timestamp()
```

### fromtimestamp

```python
from datetime import datetime
datetime.fromtimestamp(1711430000)
```
