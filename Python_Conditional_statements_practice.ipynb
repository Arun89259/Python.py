```python
# 1. Take two numbers from the user and print their sum, difference, product, division, and floor division.

a = int(input("enter first number: "))
b = int(input("enter second number: "))

print(a,b)
print("sum = ", a + b)
print("Difference = ", a - b)
print("Product = ", a*b)
print("Division", a / b)
print("Floor", a//b)
```
